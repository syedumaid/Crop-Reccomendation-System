```python
train.loc[train['soil_type']==' Laterite or red or alluvial', 'soil_type'] = 0
train.loc[train['soil_type']=='Clayey or Latterite', 'soil_type'] =1
train.loc[train['soil_type']==' loamy or sandy or Clayey', 'soil_type'] =2
train.loc[train['soil_type']=='sandy ', 'soil_type'] = 3
train.loc[train['soil_type']=='Black or Red or Loamy', 'soil_type'] =4
train.loc[train['soil_type']=='sandy or red or lateritic  or Clayey', 'soil_type'] = 5
train.loc[train['soil_type']=='Clayey or sandy', 'soil_type'] =6
train.loc[train['soil_type']=='Sandy', 'soil_type'] = 7
train.loc[train['soil_type']=='Alluvial or Black', 'soil_type'] =8
train.loc[train['soil_type']=='Clayey or Alluvial or Black', 'soil_type'] = 9
train.loc[train['soil_type']=='Clayey', 'soil_type'] =10
train.loc[train['soil_type']=='Sandy or Black', 'soil_type'] =11
train.loc[train['soil_type']=='Clayey or Alluvial', 'soil_type'] =12
train.loc[train['soil_type']=='Loamy or Black', 'soil_type'] =13
train.loc[train['soil_type']=='Black or Red', 'soil_type'] =14
```

```python
X=train[["soil_type","Area"]]
y=train["Crop"]
```

```python
y
```

```python
from sklearn.ensemble import RandomForestClassifier

Classifier = RandomForestClassifier(n_estimators=15, random_state=0)
Classifier.fit(X,y)
```

```python
print(Classifier.score(X,y))
```

```python
predictionss=Classifier.predict([[1254.0,0]])
pred=format((predictionss[0]))
```
